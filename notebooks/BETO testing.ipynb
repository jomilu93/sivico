```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import pandas as pd
```

```python
df = pd.read_csv(r'/home/ismaelod/code/jomilu93/sivico/data/senators_data.csv',converters={'initiative_list': pd.eval})
```

```python
df
```

```python
tokenizer = AutoTokenizer.from_pretrained('mrm8488/bert2bert_shared-spanish-finetuned-summarization')
model = AutoModelForSeq2SeqLM.from_pretrained('mrm8488/bert2bert_shared-spanish-finetuned-summarization')
```

```python
ejemplo = df['initiative_list'][35]
```

```python
concatenado = 'Iniciativa: '.join(ejemplo)
```

```python
ejemplo
```

```python
inputs = tokenizer.encode('RESUMEN: ' + ejemplo[0], return_tensors='pt')
```

```python
resumen = model.generate(
 inputs, 
 max_length=400, 
 min_length=100, 
 length_penalty=2.0, 
 num_beams=4, 
 early_stopping=True)
print(resumen)
```
